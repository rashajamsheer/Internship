2. Write a python program to display IMDB’s Top rated 50 movies’ data (i.e. name, rating, year of release) 
   and make data frame.

```python
page = requests.get("https://www.imdb.com/chart/top/")
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
title = []
rate =[]
year = []

for i in soup.find_all('td',class_='titleColumn'):
    # Remove new line characters and single quotes from the text
    title_text = i.get_text().replace('\n', '').replace("'", "")
    # Extract the title name from the text
    title_name = title_text.strip().split('.')[1].split('(')[0].strip()
    title.append(title_name)
    
for i in soup.find_all('span', class_='secondaryInfo'):
    year.append(i.get_text().strip('()'))

for i in soup.find_all('td',class_='ratingColumn imdbRating'):
    rate.append(i.get_text().replace('\n',''))    
```

```python
len(title)
```

```python
len(year)
```

```python
len(rate)
```

```python
df = pd.DataFrame({"Title":title[:50],"Year":year[:50],"Rate":rate[:50]})
df
```

5. Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape and make data frame

a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating

```python
page = requests.get("https://www.icc-cricket.com/rankings/mens/team-rankings/odi")
page
```

```python
soup=BeautifulSoup(page.content)
```

```python
team = []
matches = []
point = []
rating = []

for i in soup.find_all('span', class_='u-hide-phablet'):
    team.append(i.get_text().replace('\n',''))

for i in soup.find_all('td', class_='table-body__cell u-center-text',limit=18):
    matches.append(i.get_text().replace('\n',''))

for i in soup.find_all('td', class_='table-body__cell u-center-text',limit=18):
    point.append(i.get_text().replace('\n',''))

for i in soup.find_all('td', class_='table-body__cell u-text-right rating',limit=9):
    rating.append(i.get_text().replace('\n',''))

banner_row = soup.find('tr', class_='rankings-block__banner')
matches.insert(0, banner_row.find('td', class_='rankings-block__banner--matches').get_text())
point.insert(0, banner_row.find('td', class_='rankings-block__banner--points').get_text())
rating.insert(0, banner_row.find('td', class_='rankings-block__banner--rating u-text-right').get_text())
```

```python
len(team[:10])
```

```python
len(point[:len(point):2])
```

```python
match = matches[0:1] + matches[1:len(matches):2]
len(match)
```

```python
rating = [r.strip() for r in rating]
len(rating)
```

```python
df = pd.DataFrame({"TEAM":team[:10],"POINT":point[:len(point):2],"MATCHES":match,"RATING":rating})
df
```

7) Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world and 
make data frame
i) Headline
ii) Time
iii) News Link

```python
page = requests.get("https://www.cnbc.com/world/?region=world")
page
```

```python
soup=BeautifulSoup(page.content)
```

```python
headline=[]
time=[]
link=[]
for i in soup.find_all('a', class_='LatestNews-headline'):
    headline.append(i.get_text().replace('\n',''))

for i in soup.find_all('time',class_='LatestNews-timestamp'):
    time.append(i.get_text().replace('\n',''))    
    
for i in soup.find_all('a', class_='LatestNews-headline'):
    link.append(i.get('href'))
  
```

```python
df = pd.DataFrame({"HEADLINE":headline,"TIME":time,"NEWS LINK":link})
df
```

9) Write a python program to scrape mentioned details from dineout.co.in and make data frame

i) Restaurant name
ii) Cuisine
iii) Location
iv) Ratings
v) Image URL

```python
page = requests.get("https://www.dineout.co.in/delhi-restaurants/buffet-special")
page
```

```python
soup=BeautifulSoup(page.content)
```

```python
name=[]
cuisine=[]
location=[]
rating=[]
link=[]
for i in soup.find_all('div', class_='restnt-info cursor'):
    name.append(i.get_text().replace('\n',''))

for i in soup.find_all('span',class_='double-line-ellipsis'):
    cuisine.append(i.get_text().split('|')[1])   
    
for i in soup.find_all('div',class_='restnt-loc ellipsis'):
    location.append(i.get_text().replace('\n',''))    
    
for i in soup.find_all('div',class_='restnt-rating rating-4'):
    rating.append(i.get_text().replace('\n',''))  
    
for i in soup.find_all('img', class_='no-img'):
    link.append(i.get('data-src'))
  
```

```python
df = pd.DataFrame({"RESTAURANT NAME":name,"CUISINE":cuisine,"LOCATION":location,"RATING":rating,"IMAGE URL":link})
df
```
