```python
import selenium
import pandas as pd
import time
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException,ElementNotInteractableException  
```

# 1

```python
driver = webdriver.Chrome("chromedriver.exe")
url = 'https://www.naukri.com/'
driver.get(url)
```

```python
# enter 'data analyst' in the designation field
search_job=driver.find_element_by_class_name('suggestor-input ')
search_job.send_keys("Data Analyst")
# eneter 'Bangalore' in the location field
search_loc =driver.find_element_by_xpath("//input[@placeholder='Enter location']")
search_loc.send_keys("Bangalore")
# for clicking search button
search_btn = driver.find_element_by_xpath("//div[@class='qsbSubmit']")
search_btn.click()
```

```python
job_title=[]
job_loc=[]
company=[]
Experiance=[]
for i in range(0,4):
    for j in driver.find_elements_by_xpath("//a[@class='title ellipsis']"):
        job_title.append(j.text)
    for k in driver.find_elements_by_xpath("//li[@class='fleft br2 placeHolderLi location']//span"):
        job_loc.append(k.text)
    for l in driver.find_elements_by_xpath("//a[@class='subTitle ellipsis fleft']"):
        company.append(l.text)
    for p in driver.find_elements_by_xpath('//span[@class="ellipsis fleft expwdth"]'):
        Experiance.append(p.text)
```

```python
df=pd.DataFrame({})
df["Job_Title"]=job_title[:10]
df["Job_Loc"]= job_loc[:10]
df["Company"]= company[:10]
df["Experiance"]= Experiance[:10]
df
```

```python
driver.close()
```

```python
driver = webdriver.Chrome("chromedriver.exe")
url = 'https://www.flipkart.com/'
driver.get(url)
```

```python
driver.find_element_by_xpath("//button[@class='_2KpZ6l _2doB4z']").click()
```

```python
driver.find_element_by_class_name('_3704LK').send_keys("sunglasses")
search=driver.find_element_by_class_name('L0Z3Pu').click()
```

```python
url=[]
for j in range(0,3):
    for i in driver.find_elements_by_xpath("//a[@class='_2UzuFa']"):
        url.append(i.get_attribute('href')) 
    driver.find_element_by_xpath("//a[@class='_1LKTO3']").click()
    time.sleep(4)
```

```python
brand=[]
description=[]
price=[]

for i in url[:100]:
    driver.get(i)
    time.sleep(3)
    
    try:
        j=driver.find_element_by_xpath("//span[@class='G6XhRU']")
        brand.append(j.text)
        
    except NoSuchElementException:
        brand.append('Nan')
        
    try:
        l=driver.find_element_by_xpath("//span[@class='B_NuCI']")
        description.append(l.text)
        
    except NoSuchElementException:
        description.append('Nan')
        
    try:
        k=driver.find_element_by_xpath("//div[@class='_25b18c']/div")
        price.append(k.text)
    except NoSuchElementException:
        price.append('NaN')
```

```python
df=pd.DataFrame({})
df["Brand"]=brand
df["Discription"]=description
df["Price"]=price
df
```

```python
driver.close()
```
